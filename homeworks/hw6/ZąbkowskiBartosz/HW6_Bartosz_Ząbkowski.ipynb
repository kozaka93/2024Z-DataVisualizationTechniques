```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
graph = nx.read_gml("polbooks.gml")
```

```python
palette = {
    'n': 'gray',
    'l': 'lightblue',
    'c': 'red'
}

description = {
    'neutralna': 'gray',
    'liberalna': 'lightblue',
    'konserwatywna': 'red'
}

labels = [palette[graph.nodes[node]['value']] for node in graph.nodes]

pos = nx.spring_layout(graph,k=0.2)

nx.draw_networkx_nodes(graph, pos, node_size=90, node_color=labels)

nx.draw_networkx_edges(graph, pos, width=0.5)


nx.draw_networkx_labels(graph, pos, font_size=4)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
legend_elements = [
    Patch(facecolor=color, edgecolor='black', label=key) for key, color in description.items() ]

plt.legend(handles=legend_elements, title="Rodzaj ksiki", loc="upper right",bbox_to_anchor=(1.3, 0.8), fontsize=6, title_fontsize=9)
    
plt.text(-1,-1.7,"Graf przedstawia sie ksiek o polityce Stanw Zjednoczonych opublikowanych w okresie bliskim \n wyborw prezydenckich w USA w 2004 roku i sprzedawanych przez Amazon.com. \n Poczenia midzy ksikami reprezentuj czste wsplne zakupy tych ksiek przez tych samych kupujcych.",fontsize=10)    
plt.title("Graf zalenoci pomidzy ksizkami \n ktre czsto byy kupowane w parze ")
plt.savefig("wizualizacja.png",dpi=300, bbox_inches='tight')
```
