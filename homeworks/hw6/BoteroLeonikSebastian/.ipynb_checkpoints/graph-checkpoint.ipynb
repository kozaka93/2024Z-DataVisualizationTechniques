```python
import networkx as nx

import matplotlib.pyplot as plt
import matplotlib as mpl

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("authors.csv")
df
```

```python
graph = nx.from_pandas_edgelist(df, source='author', target='coauthor', edge_attr='count')
```

```python
cc = list(nx.connected_components(graph))
cc
```

```python
centers = nx.circular_layout(nx.cycle_graph(len(cc)), scale=5)

pos = {}
for c, center in zip(cc, centers.values()):
    pos.update(nx.spring_layou(nx.subgraph(graph, c), k=20, center=center, scale=1))
    
nx.draw(graph, with_labels=True, pos=pos,
        bbox=dict(facecolor='blue', alpha=1), font_size=8, font_color="white")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 16), dpi = 300)

pos = nx.spring_layout(graph)

nx.draw_networkx_nodes(
    graph,
    pos=pos,
    node_color='lightblue',
    node_size=300,
    ax=ax
)

nx.draw_networkx_edges(
    graph,
    pos=pos
)
```
